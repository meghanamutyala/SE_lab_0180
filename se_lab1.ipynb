```python
import matplotlib.pyplot as plt
import numpy as np

# List of tuples containing values of (a, b, c)
coefficients_list = [(1, 2, 3), (0.5, 1, 2), (-0.5, -1, 3)]

# Create a figure and axis
fig, ax = plt.subplots()

# Plot each curve with different values of a, b, c
for a, b, c in coefficients_list:
    x = np.linspace(-27, 27, 500)
    y = a * (x**2) + b * x + c
    ax.plot(x, y, label=f'a={a}, b={b}, c={c}')

# Set labels and title
ax.set_xlabel("Time")
ax.set_ylabel("Temperature")
ax.set_title("Weather Modeling")

# Add a legend to the plot
ax.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Hardcoded coefficients
a_hardcoded, b_hardcoded, c_hardcoded = 1, 2, 3

# Create a figure and axis
fig, ax = plt.subplots()

# Plot the curve with hardcoded coefficients
x_hardcoded = np.linspace(-27, 27, 500)
y_hardcoded = a_hardcoded * (x_hardcoded ** 2) + b_hardcoded * x_hardcoded + c_hardcoded
ax.plot(x_hardcoded, y_hardcoded, label=f'Hardcoded: a={a_hardcoded}, b={b_hardcoded}, c={c_hardcoded}')

# User input for a, b, c
a_user = float(input("Enter a: "))
b_user = float(input("Enter b: "))
c_user = float(input("Enter c: "))

# Generate x and y values based on user input
x_user = np.linspace(-27, 27, 500)
y_user = a_user * (x_user ** 2) + b_user * x_user + c_user

# Plot the curve with user-specified coefficients
ax.plot(x_user, y_user, label=f'User Input: a={a_user}, b={b_user}, c={c_user}')

# Set labels and title
ax.set_xlabel("Time")
ax.set_ylabel("Temperature")
ax.set_title("Weather Modeling")

# Add a legend to the plot
ax.legend()

# Show the plot
plt.show()


```
